```python
# Define the categories of columns
categories = ['Ease_of_Online_Booking', 'Checkin_Service', 'Online_Boarding',
              'Gate_Location', 'Onboard_Service', 'Seat_Comfort',
              'Leg_Room_Service', 'Cleanliness', 'Food_and_Drink',
              'Inflight_Service', 'Inflight_Wifi_Service', 'Inflight_Entertainment']
# Calculate the averages for each category
category_averages = df_2[categories].mean()
# Create a bar chart
plt.figure(figsize=(10, 6))
category_averages.plot(kind='bar', color='skyblue')
plt.title('Average Ratings for Flight Experience Categories')
plt.xlabel('categories')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
#Create a histogram for the "Age" column
plt.figure(figsize=(10, 6))
plt.hist(df_2['Age'], bins=20, color='green', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
data = {
    'Age Group': ['19-30', '31-40', '41-50', '51-60', '61-70', '71+', '19-30', '31-40', '41-50', '51-60', '61-70', '71+', '19-30', '31-40', '41-50', '51-60', '61-70', '71+'],
    'Satisfaction': ['Satisfied', 'Neutral', 'Dissatisfied', 'Satisfied', 'Neutral', 'Dissatisfied', 'Satisfied', 'Neutral', 'Dissatisfied', 'Satisfied', 'Neutral', 'Dissatisfied', 'Satisfied', 'Neutral', 'Dissatisfied', 'Satisfied', 'Neutral', 'Dissatisfied']
}

df_2 = pd.DataFrame(data)

# Define labels for the age bins
age_labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71+']

# Create a cross-tabulation of Age Group and Satisfaction
cross_tab = pd.crosstab(df_2['Age Group'], df_2['Satisfaction'])

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
cross_tab.plot(kind='bar', stacked=True, color=['green', 'yellow', 'red'], edgecolor='black')
plt.title('Satisfaction by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
age_labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71+']
age_group_counts = df_2['Age Group'].value_counts().sort_index()

plt.figure(figsize=(10, 4))
age_group_counts.plot(kind='bar', color='green')
plt.title('Passenger Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
age_labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71+']

age_satisfaction_counts = df_2.groupby(['Age Group', 'Satisfaction']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
ax = age_satisfaction_counts.plot(kind='bar', stacked=True, colormap='coolwarm', edgecolor='black')
plt.title('Passenger Satisfaction by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Passengers')
plt.xticks(range(len(age_labels)), age_labels, rotation=0)
plt.legend(title='Satisfaction', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()




```
